```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
```

```python
airports = ['AGADIR','CASABLANCA','FES_SAISS','MARRAKECH','RABAT_SALE','TANGER']
df = pd.read_csv("trafic_ma_long_clean.csv")
df['date'] = df["period"].str.replace("s1", "06").str.replace("s2", "12")
df['date'] = pd.to_datetime(df['date'], format="%Y_%m")
df = df.sort_values(['airport', 'date'])
```

```python


for airport in airports:
    
    df_airport = df[df["airport"] == airport].set_index("date")
    ts = df_airport["traffic"].asfreq('6MS')  

    # Fit SARIMA (you can tune these parameters later)
    model = SARIMAX(ts, order=(1, 1, 1), seasonal_order=(1, 1, 1, 2))
    result = model.fit()

    # Forecast next 4 semesters (2 years)
    forecast = result.predict(start=len(ts), end=len(ts)+3)

    plt.figure(figsize=(10, 5))
    ts.plot(label="Historical", marker='o')
    forecast.plot(label="Forecast", marker='x')
    plt.title(f"{airport} Airport Traffic Forecast")
    plt.xlabel("date")
    plt.ylabel("traffic")
    plt.legend()
    plt.grid(True)
    plt.show()
```
